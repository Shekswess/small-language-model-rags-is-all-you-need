# Exploring results data

The idea of this notebook is to explore the results data and try to find the best RAG approach by analizing the scores of metrics.

For metrics, we are using:
- faithfulness
- answer_relevancy
- context_utilization

```python
# Importing needed libraries
import pandas as pd
```

```python
# Reading the results.csv file
dataframe = pd.read_csv("/home/bojan/Work/mixture-of-rags/results/results.csv")
```

```python
# Creating a dataframe with mean values for the scores for each experiment
dataframe_mean = dataframe.drop(columns=['trace_id']).groupby("experiment_name").mean()
```

```python
# Displaying the dataframe sorted by faithfulness by descending order
dataframe_mean.sort_values(by="faithfulness", ascending=False)
```

```python
# Displaying the dataframe sorted by answer_relevancy by descending order
dataframe_mean.sort_values(by="answer_relevancy", ascending=False)
```

```python
# Displaying the dataframe sorted by context_utilization by descending order
dataframe_mean.sort_values(by="context_utilization", ascending=False)
```

```python
# Adding mean score for each experiment(mean of all the scores)
dataframe_mean['mean'] = dataframe_mean.mean(axis=1)
```

```python
# Displaying the dataframe sorted by mean by descending order
dataframe_mean.sort_values(by="mean", ascending=False)
```
